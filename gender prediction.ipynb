# Data Prepration 

```python
# ml algortihms
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
from sklearn.metrics import plot_confusion_matrix, roc_curve, auc, roc_auc_score
```

```python
X = df.drop('gender', axis = 1)
y = df['gender']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123,
                                                   shuffle = True, stratify = y)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# Data Modeling 

```python
lr = LogisticRegression()
sgdc = SGDClassifier()
svc = SVC()
dt = DecisionTreeClassifier()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import classification_report, confusion_matrix

lr = LogisticRegression(random_state = 22)
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
y_pred = lr.predict(X_test)
y_probs = lr.predict_proba(X_test)[:, 1]

print(precision_score(y_test, y_pred))
print(recall_score(y_test, y_pred))
print(f1_score(y_test, y_pred))
print(roc_auc_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(f'ROc _AUC_SCORE {roc_auc_score(y_test, y_probs)}')
```

```python
def cross_val():
        print('Cross validaiton with 5 folds')
        print(f'ROC_AUC {round(cross_val_score(lr, X,y,cv=5, scoring = "roc_auc").mean(),2)}')
        print(f'Precision {round(cross_val_score(lr, X,y,cv=5, scoring = "precision").mean(),2)}')
        print(f'Recall {round(cross_val_score(lr, X,y,cv=5, scoring = "recall").mean(),2)}')
        print(f'F1 {round(cross_val_score(lr, X,y,cv=5, scoring = "f1").mean(),2)}')
cross_val()
```
